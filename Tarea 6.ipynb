```python
# Integración montecarlo
def int_montecarlo(f:'Función a integrar',
                   a:'Límite inferior de la integral',
                   b:'Límite superior de la integral',
                   U:'Muestra de números U~[a,b]'):
    return (b - a) / len(U) * np.sum(f(U))
```

```python
def m_estratificado_igual(a, b, B):
    Ui = np.random.uniform(a, b, B)
    Bi = np.arange(0, B)
    Ui_estra_igual = (Ui + Bi) / B
    return Ui_estra_igual
```

```python
def m_estratificado_no_igual(a, b, B):
    Ui = np.random.uniform(a, b, B)
    Bi = np.random.uniform(0, 1, B)
    Ui_estra_no_igual = (Ui + Bi) / B
    return Ui_estra_no_igual
```
